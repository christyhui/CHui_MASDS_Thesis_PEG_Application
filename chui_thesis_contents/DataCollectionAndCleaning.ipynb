If 0, then it is human written. If 1, then it is AI-generated.

```python
from dotenv import load_dotenv
import pandas as pd
from datasets import load_dataset
import kagglehub
```

```python
path = kagglehub.dataset_download("starblasters8/human-vs-llm-text-corpus")
file_path = f"{path}/data.csv"
grinberg_df = pd.read_csv(file_path, encoding = "latin1")
grinberg_df["label"] = [0 if n == "Human" else 1 for n in grinberg_df["source"]]
grinberg_df = grinberg_df.drop(columns = ["prompt_id", "text_length", "word_count"])
print(f"Dimensions: {grinberg_df.shape}")
print("\n")
print(f"Columns: {grinberg_df.columns}")
```

```python
grinberg_df.head()
```

```python
gsingh = load_dataset("gsingh1-py/train")
gsingh_df = gsingh["train"].to_pandas()
gsingh_df = pd.melt(gsingh_df, id_vars = "prompt")
gsingh_df["label"] = [0 if n == "Human_story" else 1 for n in gsingh_df["variable"]]
gsingh_df = gsingh_df.drop(columns = ["prompt"])
gsingh_df = gsingh_df.rename(columns = {"variable":"source", "value":"text"})
print(f"Dimensions: {gsingh_df.shape}")
print("\n")
print(f"Columns: {gsingh_df.columns}")
```

```python
gsingh_df.head(6)
```

```python
shahxee = load_dataset("shahxeebhassan/human_vs_ai_sentences")
shahxee_df = shahxee["train"].to_pandas()
shahxee_df["source"] = shahxee_df["label"].map({0: "human", 1: "unknown"})
print(f"Dimensions: {shahxee_df.shape}")
print("\n")
print(f"Columns: {shahxee_df.columns}")
```

```python
shahxee_df.head(6)
```

```python
anagnostou_df = pd.read_csv("https://raw.githubusercontent.com/panagiotisanagnostou/AI-GA/refs/heads/main/ai-ga-dataset.csv")
anagnostou_df = anagnostou_df.drop(columns = ["title"])
anagnostou_df = anagnostou_df.rename(columns = {"abstract":"text"})
anagnostou_df["source"] = anagnostou_df["label"].map({0: "human", 1: "GPT-3"})
print(f"Dimensions: {anagnostou_df.shape}")
print("\n")
print(f"Columns: {anagnostou_df.columns}")
```

```python
anagnostou_df.head(6)
```

```python
combined_df = pd.concat([grinberg_df, shahxee_df, anagnostou_df])
combined_df = combined_df.reset_index(drop = True)
```

```python
combined_df.to_csv("./data/all_prompts.csv")
```
